```python
import pandas as pd
import matplotlib.pyplot as plt
```


## Adam results

```python
# Read the dataset
adam_data = pd.read_csv('adam_training_results.csv')
adam_data['total_time'] = adam_data['total_time (s)']
adam_data.drop(columns=['total_time (s)'], inplace=True)
adam_data
```

```python
adam_data['label'] = adam_data['learning_rate'].astype(str) + " - " + adam_data['betas'].astype(str)+ " - " + adam_data['weight_decay'].astype(str)

sorted_adam_time_data = adam_data.sort_values(by=['total_time'], ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(adam_data['label'], sorted_adam_time_data['total_time'], color='skyblue')
plt.xlabel('Learning Rate - Betas - Weight Decay') 
plt.ylabel('Total Time for all 5 folds in seconds')
plt.title('Adam Optimizer (Total Time) vs (Each Pair)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sorted_adam_fold_acc = adam_data.sort_values(by=['avg_fold_acc'], ascending=False)

plt.bar(adam_data['label'], sorted_adam_fold_acc['avg_fold_acc'], color='red')
plt.xlabel('Learning Rate - Betas - Weight Decay') 
plt.ylabel('Average Fold Frequency')
plt.title('Average Fold Frequency vs Learning Rate and Alpha')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
